```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w=np.random.randn(1)
```

```python
w.shape
X=1.5
y=0.5
```

```python
epochs=10
lr=0.1
elist=[]
for epoch in range (epochs):
    # forward pass
    y_out=X*w
    # compute error 
    error=np.square(y-y_out)
    elist.append(error)
    # compute gradient
    gradient=-2*(y-y_out)*X
    # backPropagte error
    w= w-gradient*lr
```

```python
plt.plot(range(epochs),elist)
plt.xlabel("epoches")
plt.ylabel("error")
plt.title("epoches vs error")
```
